```python
import numpy as np
from typing import Self
from numpy.typing import NDArray
from jaxtyping import Float


type Features = Float[NDArray, 'samples features']
type Weights = Float[NDArray, 'features 1']
type Target = Float[NDArray, 'samples 1']
type Prediction = Float[NDArray, 'samples 1']

class linear_regression:
    def __init__(self):
        self._fitted: bool = False
        self._weights: Weights

    def coef_(self):
        print(self._weights)
        
    def fit(self, X: Features, y: Target, alpha: float=0.01) -> Self:
        print(type(X))
        w = np.ones(X.shape[1])
        
        X = X.copy()
        X[:, 0] = 1
        # y.reshape(-1, 1)
        print(X, w, y, sep='\n')
        def gradient(X: Features, w: Weights, y: Target) -> Weights:
            return X.T @ (X @ w - y)

        for _ in range(1000):
            w = w - alpha * gradient(X, w, y)

        self._fitted = True
        self._weights = w
        return self
        
    def predict(self, X: Features) -> Prediction:
        if self._fitted:
            return X @ self._weights
        else:
            raise OperationError('model is not fitted')
```
